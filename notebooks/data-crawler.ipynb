# Google Image Crawler
Configure and run the crawler directly from this notebook.

## Configuration

```python
KEYWORD          = "Architecture photography"
LIMIT            = 200
BASE_DIR         = str(PROJECT_ROOT / "raw_datasets" / "google_image")
REQUEST_DELAY    = 0.5
TIMEOUT          = 10
MAX_CONCURRENCY  = 10
MAX_RETRIES      = 3
BACKOFF_BASE     = 2.0
```

```python
image_filter = ImageFilter(
    size         = "large",
    image_type   = "photo",
    color        = None,
    color_type   = "color",
    usage_rights = None,
    aspect_ratio = None,
    time_range   = None,
    safe_search  = "high",
    site         = None,
)
```

## Run

```python
config = CrawlConfig(
    keyword         = KEYWORD,
    limit           = LIMIT,
    base_dir        = BASE_DIR,
    image_filter    = image_filter,
    request_delay   = REQUEST_DELAY,
    timeout         = TIMEOUT,
    max_concurrency = MAX_CONCURRENCY,
    max_retries     = MAX_RETRIES,
    backoff_base    = BACKOFF_BASE,
)

print(f"Output folder: {config.output_dir}")
```

```python
crawler = GoogleImageCrawler(config)
total   = await crawler.crawl()
print(f"\nFinished. Total downloaded: {total}")
```

# Pinterest Crawler (gallery-dl)
Downloads images from any Pinterest URL — board, pin, user profile, or search.

> **Auth note:** Pinterest requires cookies for most content. Export a `cookies.txt`
> from your browser (logged-in session) using:
> - Chrome: *Get cookies.txt LOCALLY* extension
> - Firefox: *Export Cookies* extension
>
> Then set `COOKIES_FILE` below to the file path.

## Configuration

### Supported URL formats
| Type | Example |
|---|---|
| Board | `https://www.pinterest.com/username/board-name/` |
| User (all boards) | `https://www.pinterest.com/username/` |
| Single pin | `https://www.pinterest.com/pin/123456789/` |
| Search | `https://www.pinterest.com/search/pins/?q=street style photography` |

```python
URL              = f"https://www.pinterest.com/search/pins/?q={KEYWORD}"
LIMIT            = 200
BASE_DIR         = str(PROJECT_ROOT / "raw_datasets" / "pinterest")
COOKIES_FILE     = str(PROJECT_ROOT / "cookies.txt")  # set to None if not needed

SECTIONS         = True    # include board sections
VIDEOS           = False   # download video pins
SLEEP_REQUEST    = 0.5     # seconds between requests
SLEEP_EXTRACTOR  = 1.0     # seconds before starting a new extractor
RETRIES          = 3
TIMEOUT          = 10.0
```

```python
config = PinterestConfig(
    url              = URL,
    limit            = LIMIT,
    base_dir         = BASE_DIR,
    cookies_file     = COOKIES_FILE,
    sections         = SECTIONS,
    videos           = VIDEOS,
    sleep_request    = SLEEP_REQUEST,
    sleep_extractor  = SLEEP_EXTRACTOR,
    retries          = RETRIES,
    timeout          = TIMEOUT,
)

print(f"Output folder : {config.output_dir}")
print(f"Cookies file  : {config.cookies_file}")
```

```python
crawler = PinterestCrawler(config)
total   = crawler.crawl()
print(f"\nFinished. Total downloaded: {total}")
```
