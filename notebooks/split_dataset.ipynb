```python
import shutil
import random
from pathlib import Path
```

```python
SOURCE_DIR = "/home/infinitydreamer/Desktop/SceneStructNet/raw_datasets/images"
OUTPUT_DIR = "/home/infinitydreamer/Desktop/SceneStructNet/dataset"
SPLITS = {"train": 0.8, "valid": 0.1, "test": 0.1}
SEED = 42
IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff", ".webp"}
```

```python
random.seed(SEED)

source_path = Path(SOURCE_DIR)
output_path = Path(OUTPUT_DIR)

class_folders = [f for f in source_path.iterdir() if f.is_dir()]
print(f"Found {len(class_folders)} class folders: {[f.name for f in class_folders]}")
```

```python
for split in SPLITS:
    (output_path / split).mkdir(parents=True, exist_ok=True)

summary = {}

for cls_folder in class_folders:
    cls_name = cls_folder.name
    images = [f for f in cls_folder.iterdir() if f.suffix.lower() in IMAGE_EXTENSIONS]
    random.shuffle(images)

    n = len(images)
    n_train = int(n * SPLITS["train"])
    n_valid = int(n * SPLITS["valid"])

    split_files = {
        "train": images[:n_train],
        "valid": images[n_train:n_train + n_valid],
        "test":  images[n_train + n_valid:]
    }

    for split, files in split_files.items():
        for f in files:
            dest = output_path / split / f"{cls_name}_{f.name}"
            shutil.copy2(f, dest)

    summary[cls_name] = {split: len(files) for split, files in split_files.items()}
    summary[cls_name]["total"] = n
```

```python
print(f"\n{'Class':<20} {'Total':>15} {'Train':>15} {'Valid':>15} {'Test':>15}")
print("-" * 90)
for cls_name, counts in summary.items():
    print(f"{cls_name:<20} {counts['total']:>15} {counts['train']:>15} {counts['valid']:>15} {counts['test']:>15}")
```

```python
print("\nFinal dataset structure:")
for split in SPLITS:
    total = len(list((output_path / split).iterdir()))
    print(f"  {split}/  ({total} images total)")
```
